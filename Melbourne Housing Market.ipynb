## Treinando um modelo de regresso linear
<b>Vamos agora comear a treinar o modelo de regresso. Precisamos primeiro dividir nossos dados em uma matriz X que contm os recursos para treinar, e uma matriz y com a varivel alvo, neste caso, a coluna Preo. Vamos descartar a coluna "Adress" porque s tem informaes de texto que o modelo de regresso linear no pode usar.

Arrays X e y

```python
sns.heatmap(df.corr(),cmap='coolwarm',annot=True)
plt.rcParams['figure.figsize'] = (15.5, 10.5)
```

```python
df.head()
```

```python
X = df[['Suburb','BuildingArea', 'Region', 'Landsize', 'tp', 'Rooms', 'Bedroom2', 'Bathroom', 'Car', ]]
y = df['Price']
```

## Split de treino
<b>Agora vamos dividir os dados em um conjunto de treinamento e um conjunto de testes. Vamos criar o modelo usando o conjunto de treinamento e depois usar o conjunto de testes para avaliar o modelo.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=101)
```

## Criando e treinando o modelo

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.score(X_train, y_train)
```

```python
# Printando a intercepo
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
print('Coefficients: \n', lm.coef_)
```

```python
df['Predition'] = (lm.intercept_+(df['BuildingArea']*2523.540917)+(df['Region']*28143.470581)+(df['Landsize']*(-2.112166))+(df['tp']*(-146191.032809))+(df['Rooms']*64889.719106)+(df['Bedroom2']*(-27339.722642))+(df['Bathroom']*197031.214847))
```

## Predies do nosso modelo

<b>Vamos pegar as previses em nosso conjunto de testes e ver o quo bem!

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
#Historiograma residual
sns.distplot((y_test-predictions),bins=50);
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
plt.figure(figsize=(25,8))

plt.plot(range(0, y_test.shape[0]), y_test, marker='+')
plt.plot(range(0, predictions.shape[0]), predictions, marker='o')
```
