```python
import pandas as pd
import folium

brasil = folium.Map(
    location=[-16.1237611, -59.9219642],    # Coordenadas retiradas do Google Maps
    zoom_start=4
)
```

```python
empresas_rr = empresas[empresas['state'] == 'RR']
```

```python
roraima = folium.Map(
    location=[2.81, -60],
    zoom_start=7
)

roraima
```

```python
for _, empresa in empresas_rr.iterrows():
    folium.Marker(
        location=[empresa['latitude'], empresa['longitude']],
    ).add_to(roraima)

roraima
```

```python
empresas_olinda_pe = empresas[empresas['city'] == 'OLINDA']

olinda = folium.Map(
    location=[-7.9981267, -34.9082027],
    zoom_start=13
)

for _, empresa in empresas_olinda_pe.iterrows():
    folium.Marker(
        location=[empresa['latitude'], empresa['longitude']],
    ).add_to(olinda)

olinda
```

```python
colors = {
 'AMPARO': 'pink',
 'GUADALUPE': 'blue',
 'CASA CAIADA': 'green',
 'PEIXINHOS': 'orange',
 'RIO DOCE': 'red',
 'BAIRRO NOVO': 'purple',
}

olinda = folium.Map(
    location=[-7.9981267, -34.9082027],
    zoom_start=13
)

for _, empresa in empresas_olinda_pe.iterrows():
    if empresa['neighborhood'] in colors.keys():
        folium.Marker(
    location=[empresa['latitude'], empresa['longitude']],
    popup='sou uma empresa de Olinda',
    icon=folium.Icon(color=colors[empresa['neighborhood']])
).add_to(olinda)
        
olinda
```

```python
empresas_guarapuava_pr = empresas[empresas['city'] == 'GUARAPUAVA']

gpuava = folium.Map(
    location=[-7.9981267, -34.9082027],
    zoom_start=13
)

for _, empresa in empresas_guarapuava_pr.iterrows():
    folium.Marker(
        location=[empresa['latitude'], empresa['longitude']],
    ).add_to(gpuava)

gpuava
```
